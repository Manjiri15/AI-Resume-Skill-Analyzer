```python
# ============================================
# STEP 3 — Job Role Skill Database
# ============================================

job_roles = {

    "Frontend Developer": [
        "html",
        "css",
        "javascript",
        "react",
        "bootstrap",
        "tailwind",
        "ui",
        "frontend"
    ],

    "Backend Developer": [
        "python",
        "java",
        "php",
        "nodejs",
        "mysql",
        "api",
        "backend",
        "django",
        "flask"
    ],

    "Data Analyst": [
        "python",
        "excel",
        "sql",
        "power bi",
        "tableau",
        "pandas",
        "numpy",
        "data analysis"
    ],

    "Database Analyst": [
        "sql",
        "mysql",
        "oracle",
        "mongodb",
        "database",
        "postgresql"
    ],

    "System Analyst": [
        "system analysis",
        "uml",
        "testing",
        "documentation",
        "requirement analysis"
    ],

    "Machine Learning Engineer": [
        "machine learning",
        "tensorflow",
        "keras",
        "scikit",
        "cnn",
        "deep learning",
        "opencv"
    ],

    "Cyber Security Analyst": [
        "network security",
        "ethical hacking",
        "linux",
        "firewall",
        "cyber security",
        "kali linux"
    ],

    "Android Developer": [
        "android",
        "kotlin",
        "java",
        "android studio",
        "firebase"
    ],

    "Cloud Engineer": [
        "aws",
        "azure",
        "google cloud",
        "cloud",
        "docker"
    ]
}

```

```python

# ============================================
# STEP 4 — Job Links
# ============================================

job_links = {

    "Frontend Developer":
    "https://www.linkedin.com/jobs/frontend-developer-jobs/",

    "Backend Developer":
    "https://www.linkedin.com/jobs/backend-developer-jobs/",

    "Data Analyst":
    "https://www.linkedin.com/jobs/data-analyst-jobs/",

    "Database Analyst":
    "https://www.linkedin.com/jobs/database-analyst-jobs/",

    "System Analyst":
    "https://www.linkedin.com/jobs/system-analyst-jobs/",

    "Machine Learning Engineer":
    "https://www.linkedin.com/jobs/machine-learning-engineer-jobs/",

    "Cyber Security Analyst":
    "https://www.linkedin.com/jobs/cyber-security-jobs/",

    "Android Developer":
    "https://www.linkedin.com/jobs/android-developer-jobs/",

    "Cloud Engineer":
    "https://www.linkedin.com/jobs/cloud-engineer-jobs/"
}


# ============================================
# STEP 5 — Extract Text from PDF
# Mobile Compatible
# ============================================

def extract_text_from_pdf(pdf_path):

    text = ""

    try:

        # Open PDF
        doc = fitz.open(pdf_path)

        # Read every page
        for page in doc:

            text += page.get_text()

        doc.close()

    except Exception as e:

        return f"Error reading PDF: {e}"

    return text.lower()


# ============================================
# STEP 6 — Clean Resume Text
# ============================================

def clean_text(text):

    text = re.sub(r'[^a-zA-Z0-9 ]', ' ', text)

    text = re.sub(r'\s+', ' ', text)

    return text.lower()


# ============================================
# STEP 7 — Detect Skills
# ============================================

def detect_skills(resume_text):

    detected_skills = []

    for role, skills in job_roles.items():

        for skill in skills:

            if skill.lower() in resume_text:

                detected_skills.append(skill)

    # Remove duplicates
    detected_skills = list(set(detected_skills))

    return detected_skills


# ============================================
# STEP 8 — Predict Job Roles
# ============================================

def predict_roles(skills_found):

    role_scores = {}

    for role, skills in job_roles.items():

        score = 0

        for skill in skills:

            if skill in skills_found:

                score += 1

        role_scores[role] = score

    # Sort by highest score
    sorted_roles = sorted(
        role_scores.items(),
        key=lambda x: x[1],
        reverse=True
    )

    return sorted_roles


# ============================================
# STEP 9 — Resume Analysis Function
# ============================================

def analyze_resume(pdf_file):

    if pdf_file is None:

        return "Please upload a PDF resume."

    # Extract text
    resume_text = extract_text_from_pdf(pdf_file)

    # Clean text
    cleaned_text = clean_text(resume_text)

    # Detect skills
    skills = detect_skills(cleaned_text)

    # Predict roles
    roles = predict_roles(skills)

    # ====================================
    # Build Output
    # ====================================

    output = ""

    output += "==================================\n"
    output += " DETECTED SKILLS\n"
    output += "==================================\n\n"

    if len(skills) == 0:

        output += "No skills detected.\n"

    else:

        for skill in skills:

            output += f"• {skill.title()}\n"

    output += "\n\n"

    output += "==================================\n"
    output += " SUITABLE JOB ROLES\n"
    output += "==================================\n\n"

    found_role = False

    for role, score in roles:

        if score > 0:

            found_role = True

            output += f"{role}\n"
            output += f"Match Score: {score}\n"

            if role in job_links:

                output += f"Apply Here:\n{job_links[role]}\n"

            output += "\n--------------------------\n\n"

    if found_role == False:

        output += "No suitable roles found.\n"

    return output
```

```python
# ============================================
# STEP 10 — Create Gradio UI
# ============================================

interface = gr.Interface(

    fn=analyze_resume,

    inputs=gr.File(
        label="Upload Resume PDF",
        file_types=[".pdf"],
        type="filepath"
    ),

    outputs=gr.Textbox(
        label="Resume Analysis Result",
        lines=30
    ),

    title="AI Resume Skill Analyzer",

    description="""
    Upload your Resume PDF and the system will:

    ✔ Detect Skills
    ✔ Predict Suitable Job Roles
    ✔ Provide Job Application Links

    Mobile + Desktop Supported
    """,

    theme=gr.themes.Soft()
)


# ============================================
# STEP 11 — Launch Application
# ============================================

interface.launch(share=True)
```
